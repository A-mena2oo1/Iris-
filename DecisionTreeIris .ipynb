```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import tree
```

```python
df = pd.read_csv("C:/Users/MT/Downloads/Iris (1).csv")
```

```python
df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(['Species','Id'],axis="columns"), df['Species'], random_state=0)
```

```python
X_test
```

```python
clf = DecisionTreeClassifier(max_depth = 2, 
                             random_state = 0)
```

```python
clf.fit(X_train, Y_train)
```

```python
clf.predict(X_test)
```

```python
tree.plot_tree(clf);
```

```python
cn=df['Species'].unique()
fn=(df.drop(['Species','Id'],axis="columns")).columns
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,3), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
y_pred = clf.predict(X_test) 

acc_test=accuracy_score(Y_test,y_pred)
acc_train = clf.score(X_train, Y_train)
```

```python
print("DT Accuracy train is: {:.2f}".format(acc_train*100))
print("DT Accuracy test is: {:.2f}".format(acc_test*100))
```
